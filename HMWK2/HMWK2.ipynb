```python
import asyncio
from pynq.overlays.base import BaseOverlay
import pynq.lib.rgbled as rgbled
import time
import threading
import random

base = BaseOverlay("base.bit")

#ACCESS THE LIGHTS = philosopher
#Turn the lights on and off
#Make the procedures
#Solve



def Eat(led, saul):
    if saul:
        for i in range(random.randint(10,20)):
            led.on()
            time.sleep(.1)
            led.off()
            time.sleep(.1)
    else:
        for i in range(random.randint(10,20)):
            led.write(0x3)
            time.sleep(.1)
            led.write(0x0)
            time.sleep(.1)

def Starve():
    time.sleep(2)
    
def Nap(led, saul):
    if saul:
        for i in range(random.randint(1,5)):
            led.on()
            time.sleep(1)
            led.off()
            time.sleep(1)
    else:
        for i in range(random.randint(1,5)):
            led.write(0x3)
            time.sleep(1)
            led.write(0x0)
            time.sleep(1)
    
def Philosopher(left_fork, right_fork, my_led, priority, saul, name):
    eat_count = 0
    while(True):
        resource_available = left_fork.acquire() and right_fork.acquire()
        if resource_available:
            Eat(my_led, saul)
            eat_count = eat_count + 1
            left_fork.release()
            right_fork.release()
            Nap(my_led, saul)
            print(name + " have eaten: " + str(eat_count) + " times")
        else:
            Starve()

led0 = base.leds[0]  #Philosopher 1
led0.on()
time.sleep(2)
led0.off()

led1 = base.leds[1] #Philosopher 2
led1.on()
time.sleep(2)
led1.off()

led2 = base.leds[2] #Philosopher 3
led2.on()
time.sleep(2)
led2.off()

led3 = base.leds[3] #Philosopher 4
led3.on()
time.sleep(2)
led3.off()

led4 = rgbled.RGBLED(4) #Philosopher 5
led4.write(0x3)
time.sleep(.5)
led4.write(0x0)

fork1 = threading.Lock()
fork2 = threading.Lock()
fork3 = threading.Lock()
fork4 = threading.Lock()
fork5 = threading.Lock()

threads = []
t = threading.Thread(target=Philosopher, args=(fork1, fork2, led0, 1, True, "Abraham")) #add P1
threads.append(t)
t.start()

t = threading.Thread(target=Philosopher, args=(fork2, fork3, led1, 2, True, "Carter")) #add P2
threads.append(t)
t.start()

t = threading.Thread(target=Philosopher, args=(fork3, fork4, led2, 3, True, "Sam")) #add P3
threads.append(t)
t.start()

t = threading.Thread(target=Philosopher, args=(fork4, fork5, led3, 4, True, "Timmy")) #add p4
threads.append(t)
t.start()

t = threading.Thread(target=Philosopher, args=(fork5, fork1, led4, 5, False, "John")) #add P5
threads.append(t)
t.start()


```
